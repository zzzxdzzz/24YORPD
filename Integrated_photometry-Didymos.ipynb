            r_a, r_b, r_c = spice.bodvrd(target, 'RADII', 3)[1]

## Preprocessing of KRC Model

The temperature models are provided by T.N. Titus using KRC.  The models are in a multidimentional array, with 48 (Local Time) x 19 (Latitudes) x 100 (depths) x 6 (thermal inertia) x 4 (emissivity).  Parameters listed below, based on Titus's email on 2022-05-23.
* Local time starts at midnight in increments of 0.5 hours.
* Latitudes start at -90 and proceed to +90 in 10 degree increments.
* Depths are in fractions of a skin depth specified in FITS extensions.
* Thermal Inertia starts at 20, 40, 80, 160, 320, 640 tiu.
* Emissivity 1.00, 0.9, 0.8, 0.7.

The mapping from temperature model to Juno's simulated images using triangulate plate shape model is performed with IDL function 'tpm_sph2plt.pro'.  This function requires temperature model to be in a 3D array of (N, D, M) in N local solar times starting from local noon, D depths, and M latitudes.  Therefore, the original model needs to be processed to format to comply with this format.

In this step, the raw temperature models are processed and broken into individual temperature files that each corresponding to one combination of thermal inertia and emissivity.  Each temperature profile will have the format complying with the requirement by 'tpm_sph2plt.pro'.

The output are saved in file names `'KRC/tref_gamma{:d}_emis{:.1f}.fits'.format(ti, emis)`

```python
# preprocess temperature model
#
# 1. shift local time to start at 12 if needed
# 2. break the data for different thermal inertia and emissivity combinations

from astropy.io import fits
from ALMACeres import tiu
import numpy as np
u.add_enabled_units(tiu)

# parameters from Titus:
period_orb = 4.36463 * u.year    # orbital period
rho = 1300 * u.Unit('kg/m3')    # density
c = 630 * u.Unit('J/(K kg)')   # specific heat
period_rot = 7.209531 * u.hour   # rotational period, from Viikinkoski et al. (2015), A&AL 581, 3

# thermal skin depth
askin = np.sqrt(period_orb / np.pi) * tiu / rho * c


infile = '/Users/zouxd/2024-YORPD/JYLi-code/Dimorphos_krc_temps.fits'

with fits.open(infile) as f_:
    temp = f_[0].data * u.K
    lst = f_[1].data * u.hour
    lat = np.squeeze(f_[2].data) * u.deg
    ti = np.squeeze(f_[3].data) * tiu
    emis = np.squeeze(f_[4].data)
    dep = np.squeeze(f_[5].data) * u.m    

# check and adjust starting local time
lst = lst % (24 * u.h)
s = len(lst) - np.where(np.isclose(lst, 12 * u.h))[0]
if s != 0:
    lst = np.roll(lst, s)
    temp = np.roll(temp, s, axis=-1)

if not np.isclose(lst[0], 12 * u.hour):
    raise ValueError('something went wrong with shifting arrays')

# make the format in compliance with the requirement by IDL routine 'tpm_sph2plt.pro'
# which requires the array to have the shape (lst, depth, latitude) when read in by IDL.
# note that the array indexing order in IDL is the opposite as in Python.
temp = np.moveaxis(temp, 0, 3)  # move the depth axis to between lst and latitude

for i, e in enumerate(emis):
    for j, t in enumerate(ti):
        # slice the temperature model
        tref = temp[i, j]

        # save reference temeprature model
        hdu0 = fits.PrimaryHDU(tref.to_value('K'))
        hdu0.header['bunit'] = 'K'
        hdu0.header['ti'] = t.to_value('tiu'), 'thermal inertia [tiu]'
        hdu0.header['emiss'] = e, 'emissivity'
        hdu0.header['rho'] = rho.to_value('kg/m3'), 'density [kg/m**3]'
        hdu0.header['c'] = c.to_value('J/(K kg)'), 'specific heat [J/kg]'
        hdu0.header['p_orb'] = period_orb.to_value('year'), 'orbital period [year]'
        hdu0.header['p_rot'] = period_rot.to_value('hour'), 'rotational period [hour]'
        hdu0.header['a_skin'] = (np.sqrt(period_orb / np.pi) * t / (rho * c)).to_value('m'), 'annual thermal skin depth [m]'
        hdu0.header['d_skin'] = (np.sqrt(period_rot / np.pi) * t / (rho * c)).to_value('m'), 'diurnal thermal skin depth [m]'
        hdu1 = fits.ImageHDU(lst.to_value('hour'), name='lst')
        hdu1.header['bunit'] = 'hour'
        hdu2 = fits.ImageHDU(dep[:, j].to_value('m'), name='depth')
        hdu2.header['bunit'] = 'm'
        hdu3 = fits.ImageHDU(lat.to_value('deg'), name='lat')
        hdu3.header['bunit'] = 'deg'
        outfile = '/Users/zouxd/2024-YORPD/JYLi-code/KRC/tref_gamma{:.0f}_emis{:.1f}.fits'.format(t.value, e)
        fits.HDUList([hdu0, hdu1, hdu2, hdu3]).writeto(outfile, overwrite=True)
```

## Calculate Simulated Image

Simulated images are calculated with radiative transfer models.  The simulated image cubes are calculated with IDL script [temp_simu.pro](temp_simu.pro) from the temperature models.  FITS files are saved in subdirectory 'temp_maps'.

```python
from glob import glob

simudir = glob('./temp_maps/*')
imfiles = glob(simudir[0] + '/tempmap_???.fits')  # for simulated images, use '/tempmap_simu??.fits'
print(imfiles[0])
print(len(imfiles))
```

```python
# apply radiative transfer model to simulated images
#
# use script 'rt_model.py' to do manual parallel computation.
#


from glob import glob
from astropy.modeling.physical_models import BlackBody
from astropy.io import fits
from ALMACeres import Layer, Surface
import time

# observation frequency
freq = 233 * u.GHz
wave = freq.to_value('m', u.spectral())

# index of refraction
nn = np.linspace(1.001, 2.5, 11)
# loss tangent
loss = np.logspace(-4, 0, 9)

simudir = glob('./temp_maps/*')

# fits keys to propagate
keys = ['ti', 'emiss', 'rho', 'c', 'p_orb', 'p_rot', 'a_skin', 'd_skin', 'long']

# loop through all subdirectories, corresponding to different TIs
for d in simudir:
    print('directory {}: starts on '.format(os.path.basename(d)), end='')
    print(time.strftime("%b-%d %H:%M:%S", time.localtime()))
    
    t00 = time.time()
    
    # collect image files
    imfiles = glob(d + '/tempmap_???.fits')  # for lightcurve simulation
    # imfiles = glob(d + '/tempmap_simu??.fits')  # for ALMA images
    
    # loop through all images
    for f in imfiles:
        print('    simu image {}: '.format(os.path.basename(f)), end='')
        t0 = time.time()
        
        # load model
        with fits.open(f) as f_:
            intensity = BlackBody(f_[0].data * u.K)(freq).to_value('Jy/arcsec2')
            zz = f_[1].data
            emi = f_[2].data
            hdr = f_[0].header
        
        # loop through dielectric parameter space
        ims = np.zeros(nn.shape + loss.shape + intensity.shape[1:])
        for i in range(intensity.shape[-2]):
            for j in range(intensity.shape[-1]):
                if intensity[0, i, j] <= 0:
                    continue
                layer = Layer(1.5, 1e-2, profile=[zz, intensity[:, i, j]])
                surface = Surface(layer)
                for k, n in enumerate(nn):
                    for t, l in enumerate(loss):
                        surface.layers[0].n = n
                        surface.layers[0].loss_tangent = l
                        ims[k, t, i, j] = surface.emission(emi[i, j], wave)

        # save simulated images
        tmp = f.split('_')
        outfile = '_'.join(tmp[:-1] + ['rt'] + [tmp[-1]])
        hdu = fits.PrimaryHDU(ims.astype('float32'))
        hdu.header['bunit'] = 'Jy/arcsec2'
        for k in keys:
            hdu.header[k] = hdr[k], hdr.comments[k]
        hdu1 = fits.ImageHDU(nn.astype('float32'), name='refidx')
        hdu2 = fits.ImageHDU(loss.astype('float32'), name='loss')
        fits.HDUList([hdu, hdu1, hdu2]).writeto(outfile, overwrite=True)
        
        print('file processing time {:.2f} min'.format((time.time() - t0) / 60))
    print('directory processing ends on ', end='')
    print(time.strftime("%b-%d %H:%M:%S", time.localtime()), end=', ')
    print('total processing time {:.2f} min'.format((time.time() - t00) / 60))
```

## Average Brightness Temperature

This section calculates disk-averaged brightness temperature from models.

```python
# calculate disk-averaged brightness temperature and saveb

from glob import glob
from astropy.io import ascii

# if True, then calculate rotational model
# if False, calculate simulated images for actual data
rotational_model = True


if rotational_model:
    # full rotation model
    lon = range(0, 360, 5)  # longitude
    filename_pattern = '/*_rt_???.fits'  # input image name pattern
    outfile = 'juno_integrated_model_refined_20231026.fits'  # output file
else:
    # simulated images for actual data
    lon = ascii.read('data/juno_data_geom.csv')['SOLon']  # for actual data
    filename_pattern = '/*_rt_simu??.fits'  # # input image name pattern
    outfile = 'juno_integrated_model_data_refined_20231026.fits' # output file


freq = 233 * u.GHz  # observation frequency
pxlscl = 2 * u.mas  # pixel scale of simu images

simudir = glob('./temp_maps/*')

flux_all = []
tb_all = []
gamma= []
for d in simudir:
    print('processing {}'.format(os.path.basename(d)))
    
    w1 = d.find('gamma')
    w2 = d.find('_', w1)
    gamma.append(float(d[w1 + len('gamma'):w2]))
    simufiles = glob(d + filename_pattern)
    nlon = len(simufiles)
    flux_d = []  # fluxes of all files in one directory
    tb_d = []  # brightness temperature of all files in one directory
    A_d = []  # cross-section
    for f in simufiles:
        with fits.open(f) as f_:
            im = f_[0].data * u.Unit(f_[0].header['bunit'])
            nn = f_['refidx'].data
            loss = f_['loss'].data
        flux = (im.sum(axis=-1).sum(axis=-1) * pxlscl**2).to('mJy')
        A = (im[0, 0] > 0).astype(int).sum() * pxlscl**2
        tb = flux.to('K', u.brightness_temperature(freq, A))
        # print('{:.2f}, {:.2f}, {:.4f}'.format(flux[5, 4], tb[5, 4], A.to('arcsec2')))
        flux_d.append(flux)
        tb_d.append(tb)
        A_d.append(A)
    flux_all.append(u.Quantity(flux_d))
    tb_all.append(u.Quantity(tb_d))

# reorganize output
flux_all = u.Quantity(flux_all).reshape(8, 5, nlon, len(nn), len(loss))
tb_all = u.Quantity(tb_all).reshape(8, 5, nlon, len(nn), len(loss))
A_d = u.Quantity(A_d)
gamma = np.reshape(gamma, (8, 5))[:, 0]
emis = [0.6, 0.7, 0.8, 0.9, 1.0]

# order by increasing TI
st = gamma.argsort()
gamma = gamma[st]
flux_all = flux_all[st]
tb_all = tb_all[st]


# save results
hdu0 = fits.PrimaryHDU(tb_all.value)
hdu0.header['bunit'] = str(tb_all.unit)
hdu0.header['ax1_name'] = 'loss', 'loss tangent'
hdu0.header['ax2_name'] = 'refidx', 'index of refraction'
hdu0.header['ax3_name'] = 'lon', 'longitude'
hdu0.header['ax4_name'] = 'emis', 'emissivity'
hdu0.header['ax5_name'] = 'ti', 'thermal inertia'
hdu1 = fits.ImageHDU(flux_all.value, name='flux')
hdu1.header['bunit'] = str(flux_all.unit)
hdu2 = fits.ImageHDU(loss, name='loss')
hdu3 = fits.ImageHDU(nn, name='refidx')
hdu4 = fits.ImageHDU(lon, name='lon')
hdu4.header['bunit'] = 'deg'
hdu5 = fits.ImageHDU(emis, name='emis')
hdu6 = fits.ImageHDU(gamma, name='ti')
hdu6.header['bunit'] = 'tiu'
fits.HDUList([hdu0, hdu1, hdu2, hdu3, hdu4, hdu5, hdu6]).writeto(outfile, overwrite=True)
```

### Test how to calculate means

Some parameters are sampled in logarithmic scale.  Not sure how to calculate the weighted mean.  We are discussing three possible approaches:

1. Take simple average
1. Average on the index
1. Average according to the distribution function

I think the last approach is probably the most reasonable based on the mathematical concept of mean:

Assuming the parameter has a distribution function of $f(x)$, it's mean is,

$\bar x = \frac{\int{x f(x) dx}}{\int{f(x) dx}}$

To test this idea and compare the results from three approaches, consider a special case of a uniform distribution: we have a parameter distributed uniformly from 10 to 100.  The theoretical mean should be 55.  The three approaches to calculate the mean are tested below.

More over, if we want to use the $\chi^2$ as the weight, then the distribution function should be calculated with the $1/\chi^2$ as the weight.

The standard deviation of the mean, by the definition of standard deviation, would be,

$\sigma^2 = \frac{\int{(x - \bar x)^2 f(x) dx}}{\int{f(x) dx}}$

```python
# test the best method to calculate the mean of parameters

# generate an array of values to be tested
nn = 100  # number of samples
x1 = 10   # lower limit
x2 = 100  # upper limit
x = np.logspace(np.log10(x1), np.log10(x2), nn)  # array of values to be tested, logarithmic sampling

# theoretical mean
print('theoretical mean = {}'.format((x1 + x2)/2))

# simple mean
simple_mean = x.mean()

# index mean
index_mean = x[int(np.round(np.arange(nn).mean()))]

# linear mean
dx = x[1:] - x[:-1]
dx = (dx[1:] + dx[:-1]) / 2
dd = np.log10(dx[1]) - np.log10(dx[1])
dx = 10**(np.linspace(np.log10(dx[0]) - dd, np.log10(dx[-1] + dd), nn))  # calculate the intervals dx
linear_mean = (x * dx).sum() / dx.sum()


print('simple mean = {:.2f}'.format(simple_mean))
print('index mean = {:.2f}'.format(index_mean))
print('linear mean = {:.2f}'.format(linear_mean))
```

```python
# fix index of refraction and fit all photometry points in **flux**
#
# The problem with the current solution is that the TI is two low for the best-fit index of refraction.
# Low TI indicates high porosity, which in turn indicates low index of refraction.
# This test tests how strong the constraints are for low TI and high index of refraction.

from astropy.io import ascii
from astropy.table import Table, hstack

weighted = True
n = 10   # fit the first n best solutions
nn_fixed = 1.45

# load data
obs_data = ascii.read('hunter_etal_table.ecsv')
obs_geom = ascii.read('data/juno_data_geom.csv')
data = hstack([obs_data, obs_geom])

# load model
with fits.open('juno_integrated_model_data_refined_20231026.fits') as f_:
    flux_model = f_[1].data * u.Unit(f_[1].header['bunit'])
    loss = f_['loss'].data
    nn = f_['refidx'].data
    lon = f_['lon'].data
    emis = f_['emis'].data
    ti = f_['ti'].data

flux_model = np.moveaxis(flux_model, 2, -1)  # move image axis to the last
flux_model = flux_model[:, :, np.isclose(nn, nn_fixed, rtol=1e-2)]

if weighted:
    chisq = ((flux_model - data['flux'])**2/data['flux_err']**2).sum(axis=-1)
else:
    chisq = ((flux_model - data['flux'])**2).sum(axis=-1)
rms_fit = []
chi_sq = []
ti_fit = []
emis_fit = []
loss_fit = []
index = []
for i in range(n):
    best = np.unravel_index(chisq.argmin(), chisq.shape)
    chi_sq.append(chisq[best])
    rms_fit.append(np.sqrt(((flux_model[best] - data['flux'])**2).sum()/len(data)))
    ti_fit.append(ti[best[0]])
    emis_fit.append(emis[best[1]])
    loss_fit.append(loss[best[3]])
    if weighted:
        chisq[best] = 1e5
    else:
        chisq[best] = 1e5 * u.mJy**2
    best = list(best)
    best[1] = emi_index[0]  # set to index of the fixed `emis_fixed`
    index.append(best)

# save best fit parameters
results = Table([u.Quantity(chi_sq)/len(data), rms_fit, u.Quantity(rms_fit)/data['flux'].mean(), ti_fit, emis_fit, loss_fit], names=['red.chisq', 'rms', 'rel.rms', 'ti', 'emis', 'loss'])
#results['chisq'].format='.2f'
results['red.chisq'].format='.2f'
results['rms'].format='.2f'
results['rel.rms'].format='.4f'
results.write('model_{}fitting_results_flux_refined_nn{:.1f}.ecsv'.format('weighted_' if weighted else '', nn_fixed), overwrite=True)

# save best fit index
np.save('model_{}fitting_index_flux_refined_nn{:.1f}.npy'.format('weighted_' if weighted else '', nn_fixed), index)

results.pprint(max_lines=-1, max_width=-1)
```

## Effective Emissivity

The effective emissivity is calculated here as the ratio of the measured flux to the modeled flux from the surface.

7/5/2023:

The concept of "effective emissivity" calculated here is confusing.  The traditional way of calculating the "mm-wavelength effective emissivity" is to use a spherical shape of the same effective size and a non-rotating body, i.e., STM, insolated and viewed at the same geometry to calculate the expected flux (e.g., [Redman et al. 1998](https://iopscience.iop.org/article/10.1086/300495)).  But the intention here is to study the rotational variations in thermal and dielectric properties after taking into account the effect of shape.  Therefore, we cannot use the traditional concept of effective emissivity here.

The go-around is to not invoke effective emissivity, but rather just use deviation from the best-fit model in the total flux.  The notations in the following cells are not changed.  But the interpretation should not be effective emissivity.

```python
# effective emissivity with NEATM model

from astropy.io import ascii
from jylipy.thermal import NEATM, STM
from jylipy import Measurement

obs = ascii.read('hunter_etal_table.ecsv')

# mean observed flux and error
flux = Measurement(obs['flux'], error=obs['flux_err'])
flux = flux.mean()

# Juno NEATM parameter based on Erin and Woodward (2010) https://iopscience.iop.org/article/10.1088/0004-6256/140/4/933
# albedo = 0.191 +/- 0.018
# diameter = 262.012 +/- 12.047
# eta = 0.999 +/- 0.075
eta = 0.999
rad = 262.012 / 2 * u.km
alb = 0.191

juno_neatm = NEATM(2.07 * u.au, rad, albedo=alb*0.3, emissivity=.95, beaming=eta)
flux_neatm = juno_neatm.fluxd(233 * u.GHz, 1.97 * u.au, unit='mJy')
emiss_neatm = flux.data / flux_neatm.value
emiss_err_neatm = flux.error / flux_neatm.value
print('NEATM-based effective emissivity: {:.4f} +/- {:.4f}'.format(emiss_neatm, emiss_err_neatm))

# radius from Viikinkoski et al. (2015)
rad_15 = 249. / 2 * u.km

juno_stm = STM(2.07 * u.au, rad_15, albedo=0.09, emissivity=.95)
flux_stm = juno_stm.fluxd(233 * u.GHz, 1.97 * u.au, unit='mJy')
emiss_stm = flux.data / flux_stm.value
emiss_err_stm = flux.error / flux_stm.value
print('STM-based effective emissivity: {:.4f} +/- {:.4f}'.format(emiss_stm, emiss_err_stm))
```
